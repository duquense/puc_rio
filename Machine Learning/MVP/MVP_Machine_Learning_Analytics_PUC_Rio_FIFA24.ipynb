Nesta etapa, iremos fazer tratamentos no dataset para permitir uma maior eficácia do nosso modelo.

1. Primeiramente, faremos um tratamento de missings e nulos, para garantir a integridade de todos os atributos.

## 3.1 Tratamento de Nulos e Missings

```python
# verificando nulls no dataset
dataset.isnull().sum()
```

Apenas a coluna 'GK' apresenta nulos. Poderíamos substituir os nulos por 0 para caracterizar a inaptidão de um atleta de linha para atuar como goleiro.

Porém, para evitar distorções nos algoritmos de classificação, julgamos melhor eliminar esta coluna, e rodar os modelos apenas com as colunas válidas para todos os atletas.

```python
# removendo a coluna 'GK'
dataset = dataset.drop(['GK'], axis=1)

# exibindo visualização matricial da nulidade do dataset
ms.matrix(dataset)
```

Agora sim, temos um dataset 100% preenchido, sem valores nulos.

Vamos seguir com o pré processamento na etapa de *encoding*.

## 3.2 One Hot Encoding

Vamos utilizar a técnica de *one-hot encoding* para transformar variáveis categóricas em binárias.


Começamos com o encoding do atributo de pé preferido do atleta:

```python
# One-Hot Encoding através do SKLearn

# Carrego o atributo que quero fazer o encoding
pref_foot = dataset[['preferred foot']]

# Inicio o encoder. O método OneHotEncoder permite diversos parâmetros, vale a pena a exploração de cada um na documentação da lib
encoder = OneHotEncoder(sparse=False)

# Faço o fit_transform e adiciono os nomes das colunas a partir das categorias originais do atributo.
pref_foot_encoded = encoder.fit_transform(pref_foot)
pref_foot_encoded = pd.DataFrame(pref_foot_encoded, columns=encoder.get_feature_names_out(['preferred foot']))

# Validação do resultado
print(pref_foot_encoded)
```

##3.3 Label Encoding

Vamos agora fazer a transformação de outros dois atributos categóricos: *att work rate* e *def work rate*. Estes atributos podem assumir 3 valores: *low, medium* e *high*.

Como estes valores apresentam uma clara ordenação (*high* > *medium* > *low*), optamos por um *label encoding* ao invés do *one-hot encoding*.

Para isso, podemos usar o método OrdinalEncoder do SciKitLearn:

```python
# Definindo a coluna a ser feito o encoding
att_work_rate = dataset[['att work rate']]

# Definindo a ordem das categorias
ordem_work_rate = ['Low', 'Medium', 'High']

# Criando uma instância do OrdinalEncoder com a ordem especificada
encoder = OrdinalEncoder(categories=[ordem_work_rate])
encoder.fit(att_work_rate)

# Faço o fit_transform e adiciono os nomes das colunas a partir das categorias originais do atributo.
att_work_rate_encoded = encoder.fit_transform(att_work_rate)
att_work_rate_encoded = pd.DataFrame(att_work_rate_encoded, columns=['att work rate encoded'])

# Validação do resultado
print(att_work_rate_encoded)
```

```python
# Definindo a coluna a ser feito o encoding
def_work_rate = dataset[['def work rate']]

# Definindo a ordem das categorias
ordem_work_rate = ['Low', 'Medium', 'High']

# Criando uma instância do OrdinalEncoder com a ordem especificada
encoder = OrdinalEncoder(categories=[ordem_work_rate])
encoder.fit(def_work_rate)

# Faço o fit_transform e adiciono os nomes das colunas a partir das categorias originais do atributo.
def_work_rate_encoded = encoder.fit_transform(def_work_rate)
def_work_rate_encoded = pd.DataFrame(def_work_rate_encoded, columns=['def work rate encoded'])

# Validação do resultado
print(def_work_rate_encoded)
```

Agora, vamos atualizar o dataset com os campos resultantes dos *encodings*:

```python
# Adicionando as colunas criadas via one-hot encoding

dataset_encoded = pd.concat([dataset, pref_foot_encoded, att_work_rate_encoded, def_work_rate_encoded], axis=1)

# Exibição do dataset resultante
dataset_encoded.head()
```

Como pode ser visto, agora nosso dataset contém variáveis booleanas para os atributos tratados via encoding.

Porém, com isso temos um aumento da dimensionalidade do dataset. Portanto, na próxima sessão usaremos de algumas técnicas para reduzir a dimensionalidade do conjunto.

## 3.4 Feature Selection

Nesta etapa, vamos utilizar técnicas para reduzir a quantidade de atributos do nosso modelo.

Primeiro, vamos fazer o drop de algumas colunas irrelevantes para o nosso problema, como a URL do card do atleta, o seu nome, clube, nacionalidade e gênero. Serão retiradas também as colunas que foram tratadas via _encoding_.

Retiro também as colunas de 'age' e 'overall', pois a idade e o _rating_ total de uma atleta não são relevantes para a definição de sua posição de jogo.




```python
# Drop de colunas irrelevantes para o modelo ou já adicionadas via encoding.
dataset_reduzido = dataset_encoded.drop(['player id','url','gender','nation','name','club','att work rate','def work rate','preferred foot','age','overall'], axis=1)

dataset_reduzido.head()
```

Agora chegamos a um dataset mais limpo para o problema, com 38 colunas de atributos e 1 coluna target (a coluna _position_).

Vou salvar este dataset como o dataset original do problema e criar uma segunda versão do mesmo, utilizando uma técnica de seleção de atributos para reduzir ainda mais o total de colunas.

```python
# Definição de dataset original para os modelos

# movendo a coluna target 'position' para o fim do dataset
position = dataset_reduzido.pop(dataset_reduzido.columns[0])
dataset_reduzido.insert(len(dataset_reduzido.columns), position.name, position)

#renomeando o dataset
dataset = dataset_reduzido

dataset.head()
```

Utilização do método SelectKBest para manter apenas os 20 atributos de maior relação com a posição.

Salvo este dataset alternativo para comparação de modelos mais à frente.

```python
# Drop da coluna de position, que será a coluna de objetivo, Y
Xx = dataset.drop(['position'], axis=1)
yy = dataset['position']

# SelectKBest

# Seleção de 20 atributos com SelectKBest
best_var = SelectKBest(score_func=f_classif, k=20)

# Executa a função de pontuação em (Xx, yy) e obtém os atributos selecionados
fit = best_var.fit(Xx, yy)

# Reduz X para os atributos selecionados
features = fit.transform(Xx)

# Resultados
print('\nNúmero original de atributos:', Xx.shape[1])
print('\nNúmero reduzido de atributos:', features.shape[1])

# Exibe os atributos orginais
print("\nAtributos Originais:", dataset.columns[0:-1])

# Exibe as pontuações de cada atributos e os 4 escolhidas (com as pontuações mais altas)
np.set_printoptions(suppress=True, formatter={'float_kind': '{:.2f}'.format}) # 2 casas decimais
print("\nScores dos Atributos Originais:", fit.scores_)
print("\nAtributos Selecionados:", best_var.get_feature_names_out(input_features=dataset.columns[0:-1]))
```

Chamaremos este subconjunto do dataset original de "dataset_reduzido", e usaremos ele para comparação da eficácia dos algoritmos de classificação a seguir.

```python
# Obtém os nomes das colunas dos atributos selecionados
selected_columns = Xx.columns[best_var.get_support(indices=True)]

# Cria um DataFrame pandas com os atributos selecionados e seus nomes de colunas
dataset_reduzido = pd.DataFrame(features, columns=selected_columns)

# Adiciona a coluna alvo 'position' de volta ao DataFrame
dataset_reduzido['position'] = dataset['position']

dataset_reduzido.head()
```

## 4.3 Tuning do modelo

Vimos na etapa anterior que o SVM apresentou a melhor acurácia (73%) dentro de um tempo considerado aceitável (50s).

Vamos então seguir com o SVM para a resolução do problema. Faremos a seguir uma etapa de _tuning_ para otimizar os hiperparâmetros e buscar mais algum ganho de acurácia.

```python
# Tempo estimado de processamento: 50 min

# Tuning do SVM

np.random.seed(7) # definindo uma semente global para este bloco

pipelines = []
train_times = []

# Definindo os componentes do pipeline
svm = ('SVM', SVC())
standard_scaler = ('StandardScaler', StandardScaler())
min_max_scaler = ('MinMaxScaler', MinMaxScaler())

# pipelines.append(('svm-orig', Pipeline(steps=[svm])))
pipelines.append(('svm-padr', Pipeline(steps=[standard_scaler, svm])))
pipelines.append(('svm-norm', Pipeline(steps=[min_max_scaler, svm])))

# Definir os hiperparâmetros a serem otimizados e seus valores
param_grid = {
    'SVM__C': [0.1, 1, 10, 100],
    'SVM__gamma': [0.1, 0.01, 0.001, 0.0001],
    'SVM__kernel': ['rbf']
}

# Prepara e executa o GridSearchCV
for name, model in pipelines:
    start_time = time.time()  # Inicia a contagem do tempo
    grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring=scoring, cv=kfold)
    grid.fit(X_train, y_train) # dataset original
    end_time = time.time()  # Finaliza a contagem do tempo
    train_time = end_time - start_time  # Calcula o tempo necessário para treinar o modelo
    train_times.append(train_time)  # Adiciona o tempo à lista
    # imprime a melhor configuração
    print("%s - Melhor: %f usando %s - Tempo de Treinamento: %.2f segundos" % (name, grid.best_score_, grid.best_params_, train_time))
```

Após o tuning, foi possível obter cerca de 74% de acurácia com o SVM.

Consideramos este valor adequado o suficiente para o nosso problema, agora vamos testa-lo com a base de teste definida no _holdout_.

## 5.3 Function do modelo

Na célula abaixo é feito um protótipo que contém o modelo já treinado e pede um input ao usuário para a entrada de dados de um novo jogador. A function retorna na saída a posição prevista para este novo atleta.

A criação desta function permite explorar com mais facilidade o modelo, criando diferentes cenários para atletas de características distintas.

Podemos dizer que é um primeiro esforço no sentido de implantar o modelo em produção e gerar valor para os usuários.

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold
import warnings
warnings.filterwarnings("ignore")

def carregar_dataset_e_treinar_modelo(url):
    # Carrega o conjunto de dados
    colunas = ['player id', 'name', 'nation', 'club', 'position', 'age', 'overall', 'pace', 'shooting', 'passing',
               'dribbling', 'defending', 'physicality', 'acceleration', 'sprint', 'positioning', 'finishing',
               'shot power', 'long shot', 'volleys', 'penalties', 'vision', 'crossing', 'free kicks', 'curve',
               'agility', 'balance', 'reactions', 'ball control', 'composure', 'interceptions', 'heading',
               'def awareness', 'standing tackle', 'sliding tackle', 'jumping', 'stamina', 'strength',
               'aggression', 'att work rate', 'def work rate', 'preferred foot', 'weak foot', 'skill moves',
               'url', 'gender', 'GK']
    dataset = pd.read_csv(url, names=colunas, skiprows=1, delimiter=',', index_col=False)
    dataset = dataset.drop(['GK'], axis=1)

    # Encoding de atributos
    pref_foot = dataset[['preferred foot']]
    encoder = OneHotEncoder(sparse=False)
    pref_foot_encoded = encoder.fit_transform(pref_foot)
    pref_foot_encoded = pd.DataFrame(pref_foot_encoded, columns=encoder.get_feature_names_out(['preferred foot']))

    att_work_rate = dataset[['att work rate']]
    ordem_work_rate = ['Low', 'Medium', 'High']
    encoder = OrdinalEncoder(categories=[ordem_work_rate])
    encoder.fit(att_work_rate)
    att_work_rate_encoded = encoder.fit_transform(att_work_rate)
    att_work_rate_encoded = pd.DataFrame(att_work_rate_encoded, columns=['att work rate encoded'])

    def_work_rate = dataset[['def work rate']]
    encoder = OrdinalEncoder(categories=[ordem_work_rate])
    encoder.fit(def_work_rate)
    def_work_rate_encoded = encoder.fit_transform(def_work_rate)
    def_work_rate_encoded = pd.DataFrame(def_work_rate_encoded, columns=['def work rate encoded'])


    dataset_encoded = pd.concat([dataset, pref_foot_encoded, att_work_rate_encoded, def_work_rate_encoded], axis=1)
    dataset_reduzido = dataset_encoded.drop(
        ['player id', 'url', 'gender', 'nation', 'name', 'club', 'att work rate', 'def work rate',
         'preferred foot', 'age', 'overall'], axis=1)
    position = dataset_reduzido.pop(dataset_reduzido.columns[0])
    dataset_reduzido.insert(len(dataset_reduzido.columns), position.name, position)
    dataset = dataset_reduzido

    # Divisão do conjunto
    array = dataset.values
    X = array[:, 0:-1]
    y = array[:, -1]

    # Ajuste do modelo
    scaler = StandardScaler().fit(X)
    rescaledX = scaler.transform(X)
    modelo = SVC(C=100, gamma=0.001, kernel='rbf')
    modelo.fit(rescaledX, y)

    return modelo



def prever_posicao_jogo():
    # Solicitar entrada do usuário
    novo_jogador = {}
    print("Por favor, preencha as informações do novo jogador:")
    novo_jogador['name'] = input("Nome do jogador: ")
    novo_jogador['pace'] = int(input("Pace: "))
    novo_jogador['shooting'] = int(input("Shooting: "))
    novo_jogador['passing'] = int(input("Passing: "))
    novo_jogador['dribbling'] = int(input("Dribbling: "))
    novo_jogador['defending'] = int(input("Defending: "))
    novo_jogador['physicality'] = int(input("Physicality: "))
    novo_jogador['acceleration'] = int(input("Acceleration: "))
    novo_jogador['sprint'] = int(input("Sprint: "))
    novo_jogador['positioning'] = int(input("Positioning: "))
    novo_jogador['finishing'] = int(input("Finishing: "))
    novo_jogador['shot power'] = int(input("Shot Power: "))
    novo_jogador['long shot'] = int(input("Long Shot: "))
    novo_jogador['volleys'] = int(input("Volleys: "))
    novo_jogador['penalties'] = int(input("Penalties: "))
    novo_jogador['vision'] = int(input("Vision: "))
    novo_jogador['crossing'] = int(input("Crossing: "))
    novo_jogador['free kicks'] = int(input("Free Kicks: "))
    novo_jogador['curve'] = int(input("Curve: "))
    novo_jogador['agility'] = int(input("Agility: "))
    novo_jogador['balance'] = int(input("Balance: "))
    novo_jogador['reactions'] = int(input("Reactions: "))
    novo_jogador['ball control'] = int(input("Ball Control: "))
    novo_jogador['composure'] = int(input("Composure: "))
    novo_jogador['interceptions'] = int(input("Interceptions: "))
    novo_jogador['heading'] = int(input("Heading: "))
    novo_jogador['def awareness'] = int(input("Def Awareness: "))
    novo_jogador['standing tackle'] = int(input("Standing Tackle: "))
    novo_jogador['sliding tackle'] = int(input("Sliding Tackle: "))
    novo_jogador['jumping'] = int(input("Jumping: "))
    novo_jogador['stamina'] = int(input("Stamina: "))
    novo_jogador['strength'] = int(input("Strength: "))
    novo_jogador['aggression'] = int(input("Aggression: "))
    novo_jogador['weak foot'] = int(input("Weak Foot (1-5): "))
    novo_jogador['skill moves'] = int(input("Skill Moves (1-5): "))
    novo_jogador['preferred foot_Left'] = float(input("Preferred Foot Left (1.0 for yes, 0.0 for no): "))
    novo_jogador['preferred foot_Right'] = float(input("Preferred Foot Right (1.0 for yes, 0.0 for no): "))
    novo_jogador['att work rate encoded'] = float(
        input("Attacking Work Rate (0.0 for low, 1.0 for medium, 2.0 for high): "))
    novo_jogador['def work rate encoded'] = float(
        input("Defensive Work Rate (0.0 for low, 1.0 for medium, 2.0 for high): "))

    # Realizar previsão
    entrada = pd.DataFrame.from_dict(novo_jogador, orient='index').T
    array_entrada = entrada.values
    X_entrada = array_entrada[:, 1:].astype(float)
    rescaledX = scaler.transform(X_entrada)
    saida = modelo.predict(rescaledX)
    print('O modelo indica que o atleta %s deve atuar na posição de %s.' % (novo_jogador['name'], saida[0]))

# Carregar o dataset e treinar o modelo apenas uma vez
url = "https://raw.githubusercontent.com/duquense/puc_rio/main/Machine%20Learning/MVP/Bases/FIFA24/all_players.csv"
modelo = carregar_dataset_e_treinar_modelo(url)

# Exemplo de uso da função
# prever_posicao_jogo(modelo)

```
